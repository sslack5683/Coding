```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('tennis_stats.csv')
print(df.head())
```

```python
plt.subplot(2, 3, 1)
sns.regplot(data= df, x = 'Aces', y = 'Winnings')
plt.subplots_adjust(wspace = .7)
plt.subplot(2, 3, 2)
sns.regplot(data= df, x= 'Aces', y = 'ReturnGamesWon')
plt.subplot(2, 3, 3)
sns.regplot(data= df, x= 'BreakPointsOpportunities', y = 'Winnings')
plt.subplot(2, 3, 4)
sns.regplot(data= df, x= 'FirstServeReturnPointsWon', y = 'Winnings')
plt.subplots_adjust(hspace= .5)
plt.subplot(2, 3, 5)
sns.regplot(data= df, x= 'FirstServe', y= 'Winnings')
```

```python
x = df[['Aces']]
y = df[['Winnings']]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=.8, test_size=.2)

lr = LinearRegression()
lr.fit(x_train, y_train)
lr.score(x_test, y_test)

y_predict = lr.predict(x_test)
plt.scatter(y_test, y_predict, alpha = .4)
```
